# Load Cell Calibration Script

The voltage output of a load cell is designed to be highly **linear** to force applied. This means we can plot a **linear regression** of the form ``y = mx + b``, where our input (``x``) is the voltage and our output (``y``) is the load (force) applied.

Once we have our **calibration coefficients** ``m``, ``b``, we can paste these into our code to calibrate our load cell in real-time.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#TODO: replace these with your actual load/voltage values from Station 2
voltage_mV = [1250.00, 1410.00, 1570.00, 1890.00] #x
load_lbs = [0, 50, 100, 150] #y
```

```python
#fit a first-degree polynomial (linear) regression line
m,b = np.polyfit(voltage_mV, load_lbs, 1)

print(m, b)
```

```python
#plot our data
def y(x, m, b): #y = mx + b
  return np.array([(m*x_i + b) for x_i in x])

plt.plot(
    voltage_mV,
    y(voltage_mV, m, b),
    color="blue",
    label=f"y = {m:.3f}x + {b:.3f}"
)
plt.scatter(
    voltage_mV,
    load_lbs,
    color="red",
    label="Raw data"
)
plt.title("Load Cell Calibration")
plt.ylabel("Load (lbs)")
plt.xlabel("Voltage (mV)")
plt.legend()
plt.show()
```
