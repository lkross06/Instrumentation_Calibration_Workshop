# Calibration Script

The voltage output of a pressure transducer (PT) is designed to be **linear** to the pressure. This means we can plot a **linear regression** of the form ``y = mx + b``, where our input (``x``) is the voltage and our output (``y``) is the pressure measured.

Once we have our **calibration coefficients** ``m``, ``b``, we can paste these into our code to calibrate our PT in real-time.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#TODO: replace these with your actual voltage/pressure values from Station 2
voltage_mV = [248.7, 502.1, 748.9, 1003.4] #x
press_psi = [50, 100, 150, 200] #y
```

```python
#fit a first-degree polynomial (linear) regression line
m,b = np.polyfit(voltage_mV, press_psi, 1)

print(m, b)
```

```python
#plot our data
def y(x, m, b): #y = mx + b
  return np.array([(m*x_i + b) for x_i in x])

plt.plot(
    voltage_mV,
    y(voltage_mV, m, b),
    color="blue",
    label=f"y = {m:.3f}x + {b:.3f}"
)
plt.scatter(
    voltage_mV,
    press_psi,
    color="red",
    label="Raw data"
)
plt.title("PT Calibration")
plt.ylabel("Pressure (psi)")
plt.xlabel("Voltage (mV)")
plt.legend()
plt.show()
```
